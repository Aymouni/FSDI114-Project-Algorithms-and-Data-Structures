Pseudocode Challenge 2: Largest number determination

START

I have a list of numbers: 12, 4, 27, 3, 188, 50, 7

Go through each number in the list one by one
    Keep track of the biggest number seen so far
    If the current number is bigger, replace it

When done, print the biggest number

END

```python
# Assignment 1: Second Problem:

nums = [12, 4, 27, 3, 1000, 50, 7]

largest = nums[0]

for num in nums:
    if num > largest:
        largest = num

print("The largest number is:", largest)
```

```python
# Assignment 1: First Problem:

def count_even_numbers(collection):
    total = 0
    for number in collection:
        if number % 2 == 0:
            total += 1
    return total

numbers = [3, 5, 27, 20, 6, 3.5, 60]
print(count_even_numbers(numbers))
```
