## Stacks

A Stack is an ordered collection of elements where items are added and removed from one end, commonly referred to as the "top". The ordering principle is referred to as LIFO (Last in, First out)

## Challenge 1: Push and Pop

```python
# 1. Create an empty stack
stack = []

# 2. Push A, B, C onto the stack
stack.append("A")
stack.append("B")
stack.append("C")

# 3. Pop the top value and store it in a variable called top
top = stack.pop()
print("Removed value:", top)

# 4. Print the entire stack and the removed value
print("Stack:", stack)
print("Removed value:", top)

# 5. Pop the next value and print it
top = stack.pop()
print("Popped:", top)

# 6. Push D onto the stack
stack.append("D")

# 7. Pop again and print the removed value
top = stack.pop()
print("Popped:", top)

# 8. Print the entire stack
print("Final stack:", stack)
```

```python
# Simplified version (relying on built-in functions)

class Stack:
    def __init__(self):
        self.items = []

    def push(self, elements):
        self.items.append(elements)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return self.items == []
    
    def __str__(self):
        print(self.items)
```

```python
stack_obj = Stack()

stack_obj.push("A")
stack_obj.push("B")
stack_obj.push("C")

top = stack_obj.pop()
print(stack_obj.items)
print(top)
print(stack_obj.peek())
print(stack_obj.size())
```
