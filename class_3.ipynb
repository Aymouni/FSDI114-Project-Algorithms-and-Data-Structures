# Class 3 - Queue

A queue is an ordered collection of elements where items are added and removed from **opposite** ends, commonly referred to as the "Front" and "Back" or "Rear" of the queue. The ordering principle is referred to as FIFO (First in, First out) "First come first served" 

```python
# Simplified Queue (relying on built-in functions)

class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, elements):
        self.items.insert(0,elements)

    def dequeue(self):
        return self.items.pop()
    
    def __str__(self):
        return f"Queue({self.items})"

    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return self.items == []
    
```

```python
my_queue = Queue()
my_queue.enqueue("A")
my_queue.enqueue("B")
my_queue.enqueue("C")
print(my_queue)
value = my_queue.dequeue()
print(value)
print(my_queue)
```

```python
# Custom Stack Class (from scratch)

class CStack():
    class Node:
        def __init__(self, data):
            self.data = data
            self.below = None

        def __str__(self):
            return str(self.data)

    def __init__(self):
        self.top = None

    def push(self, value):
        if not self.top:
            self.top = self.Node(value)
        else:
            new_node = self.Node(value)
            new_node.below = self.top
            self.top = new_node

    # Assignment 2: Build the logid for the pop and is_empty funtions
    
    def pop(self):
        if not self.top:
            raise IndexError("Stack is Empty")      # Stack is empty, nothing to pop
        else:         
            removed = self.top.data                 # Store the top value
            self.top = self.top.below               # Move top down to the next node
            return removed

    def is_empty(self):
        return self.top == None                     # Returns True if stack is empty
    
    def __str__(self):
        return f"CStack - [{self.top}]"
    
     # Assignment 3

    def peek(self):
        if not self.top:
            raise IndexError("Stack is Empty")
        return self.top.data

    def get_size(self):
        size = 0
        current = self.top
        while current:
            size += 1
            current = current.below
        return size


# Assignment 3 - Optional 

class CustomQueue():
    class Node:
        def __init__(self, data):
            self.data = data
            self.next = None

        def __str__(self):
            return str(self.data)
        
    def __init__(self):
        self.front = None
        self.back = None

    def enqueue(self, value):
        new_node = self.Node(value)
        if not self.back:
            self.front = new_node
            self.back = new_node
        else:
            self.back.next = new_node
            self.back = new_node

    def dequeue(self):
        if not self.front:
            raise IndexError("Queue is Empty")
        removed = self.front.data
        self.front = self.front.next
        if not self.front:
            self.back = None
        return removed

    def __str__(self):
        return f"CustomQueue - [{self.front}]"
```

```python
cstack = CStack()
cstack.push("A")
cstack.push("B")
cstack.push("C")
print(cstack)
print("Peek:", cstack.peek())
print("Size:", cstack.get_size())
cstack.pop()
print("After pop:", cstack)

queue = CustomQueue()
queue.enqueue("A")
queue.enqueue("B")
queue.enqueue("C")
print(queue)
print("Dequeue:", queue.dequeue())
print("After dequeue:", queue)
```
